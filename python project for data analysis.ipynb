# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.shape
```

```python
df.dropna(inplace = True)
```

```python
#change data type
df['Amount'] = df['Amount'].astype('int')
```

# Exploratory Data Analysis

## Gender 

```python
df.columns
```

```python
sns.countplot(x = 'Gender',data = df)
```

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sales_gen
```

```python
a = sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
for bars in a.containers:
    a.bar_label(bars)
```

#### From the above 2 graphs it can be seen that most of the buyers are females(7832) compared to men(3407) and purchasing power of females(74335900) are greater than men(31913300).

## Age

```python
df.columns
```

```python
sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sales_age
ax = sns.barplot(x = 'Age Group', y = 'Amount',data = sales_age)
for bars in ax.containers:
    ax.bar_label(bars)
```

#### from the abaove graphs it can be seen that most of the buyers are of age group 26-35 and out of which females are more compared to male.

## State 

```python
ax = sns.countplot(y = 'State',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sales_state

ax = sns.barplot(x = 'Amount',y = 'State',data = sales_state)

for bars in ax.containers:
    ax.bar_label(bars)
```

#### From the above graph it can be seen that total highest amount spent on diwali slaes is from Uttar Pradesh, Maharastra, Karnataka respectively

```python
sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False)
sales_state

ax = sns.barplot(x = 'Orders',y = 'State',data = sales_state)

for bars in ax.containers:
    ax.bar_label(bars)
```

#### from the above graph it can be seen that most of the orders are from Uttar Pradesh, Maharashtra, Karnataka, Delhi respectively

```python
df.columns
```

## Marital Status 

```python
ax = sns.countplot(x = 'Marital_Status', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marit = df.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sales_marit

ax = sns.barplot(x = 'Marital_Status', y = 'Amount', data = sales_marit, hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

#### from the above graph it can be seen that most are not married and count is more in females compared to males in both married and non married categories

## Occupation 

```python
ax = sns.countplot(y = 'Occupation',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occup = df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sales_occup

ax = sns.barplot(x = 'Amount', y = 'Occupation',data = sales_occup)

for bars in ax.containers:
    ax.bar_label(bars)
```

#### From the above graphs it can be seen that most of the buyers are working in IT,Health Care, Aviation and purchasing power is also the same

## Product Category

```python
ax = sns.countplot(x = 'Product_Category', data =df)
sns.set(rc = {'figure.figsize':(35,25)})
for bars in ax.containers:
    ax.bar_label(bars)
```
